# Discovering Data Structure Definitions (DSDs) in BIS SDMX

## Overview

In the previous notebook, we discovered the list of **Dataflows** exposed by the Bank for International Settlements (BIS). A Dataflow tells us **what datasets are available**, but it does not describe **how those datasets are organized**.

This notebook explores the next layer of the SDMX Information Model — the **Data Structure Definition (DSD)**.

A DSD defines the schema of a dataset, including:

- Dimensions (e.g., Country, Frequency, Time)
- Measures (e.g., Observation Value)
- Attributes (e.g., Units, Status)
- References to codelists that define valid values

Think of a DSD as the SDMX equivalent of running:

```sql
DESCRIBE TABLE credit_statistics;
```

on a relational database.

Understanding the DSD allows us to understand a dataset **without downloading a single observation**.

````python
import sdmx
import pandas as pd

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
````

````python
# ----------------------------------------------------
# Connect to BIS SDMX API
# ----------------------------------------------------

client = sdmx.Client("BIS")
print("Connected to BIS successfully.")
````

````python
# ----------------------------------------------------
# Retrieve all dataflows
# ----------------------------------------------------


response = client.dataflow()

flows = response.dataflow

print(f"Total Dataflows : {len(flows)}")
````

````python
# ----------------------------------------------------
# Convert into DataFrame
# ----------------------------------------------------

records = []

for flow_id, flow in flows.items():

    try:
        name = flow.name.en
    except Exception:
        name = str(flow.name)

    try:
        description = flow.description.en
    except Exception:
        description = ""

    try:
        structure = flow.structure.id
    except Exception:
        structure = ""

    records.append(
        {
            "Flow ID": flow_id,
            "Name": name,
            "Description": description,
            "Structure": structure,
        }
    )

df_flows = pd.DataFrame(records)
df_flows = df_flows.sort_values("Flow ID").reset_index(drop=True)
df_flows
````

````python
dimensions = []

for dim in dsd.dimensions:

    # Linked codelist (if present)
    codelist = None

    if dim.local_representation and dim.local_representation.enumerated:
        codelist = dim.local_representation.enumerated.id

    dimensions.append(
        {
            "Order": dim.order,
            "Dimension ID": dim.id,
            "Dimension Name": str(dim.concept_identity.name),
            "Dimension Type": type(dim).__name__,
            "Codelist": codelist
        }
    )

df_dimensions = (
    pd.DataFrame(dimensions)
      .sort_values("Order")
      .reset_index(drop=True)
)

df_dimensions
````

## Explanation of df_dimensions - columns and their purpose


| Column             | Purpose                                                                          |
| ------------------ | -------------------------------------------------------------------------------- |
| **Order**          | Determines the sequence of values in an SDMX data key.                           |
| **Dimension ID**   | Used when constructing SDMX API queries.                                         |
| **Dimension Name** | Human-readable description of the dimension.                                     |
| **Dimension Type** | Indicates whether it's a regular dimension or the special time dimension.        |
| **Codelist**       | References the set of allowed values, which you'll explore in the next notebook. |

````python
first_mes = dsd.measures[0]

print(type(first_mes))

print(first_mes)

print(dir(first_mes))
````

````python
# ----------------------------------------------------
# Inspect every property
# ----------------------------------------------------

for mes in dsd.measures:
    print("=" * 80)
    print(f"Measure ID : {mes.id}")

    for attr in dir(mes):
        if attr.startswith("_"):
            continue

        try:
            value = getattr(mes, attr)
            print(f"{attr:25}: {value}")
        except AttributeError:
            pass
````

````python
rep = mes.local_representation

print(type(rep))
print(rep)
print(dir(rep))
````

````python
for attr in dir(rep):
    if attr.startswith("_"):
        continue

    try:
        print(f"{attr:25}: {getattr(rep, attr)}")
    except Exception:
        pass
````

````python
measures = []

for mes in dsd.measures:

    # Default values
    data_type = None
    max_length = None

    # Extract information from the non-enumerated representation
    if (
        mes.local_representation is not None
        and mes.local_representation.non_enumerated
    ):
        facet = mes.local_representation.non_enumerated[0]
        data_type = facet.value_type.name
        max_length = facet.type.max_length

    measures.append({
        "Measure ID": mes.id,
        "Measure Name": str(mes.concept_identity.name),
        "Measure Type": type(mes).__name__,
        "Data Type": data_type,
        "Max Length": max_length,
        "URN": mes.urn
    })

df_measures = (
    pd.DataFrame(measures)
      .reset_index(drop=True)
)

df_measures
````
